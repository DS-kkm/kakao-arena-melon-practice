```python
train_data = pd.read_json('./train.json', encoding='utf-8')
train_data = train_data[['id','tags','like_cnt']]
train_data.head()
```

```python
tags = [elem for elem in train_data["tags"]]
tag_set = set() 
for elem in tags: 
    for tag in elem: tag_set.add(tag)
len(tag_set)
```

```python
# tag_set = set(id_tags_df["tag"])
idx2tag, tag2idx = dict(), dict() 
for idx, tag in enumerate(tag_set): idx2tag[idx], tag2idx[tag] = tag, idx
idx2tag
```

```python
tag2idx
```

```python
tag2idx["시너지"], idx2tag[7972]
```

```python
tag_len, row_num = len(tag_set), len(train_data)
tag_len, row_num
```

```python
col_list = ["id"] 
col_list += [elem for elem in tag2idx]
new_df = pd.DataFrame(columns=col_list)
new_df
```

```python
# for i in range(row_num): 
for i in range(50):
    _id, tags, cnt = train_data.iloc[i] 
    row = [_id]
    tmp_tag = [0]*tag_len
    for tag in tags: 
        tmp_tag[tag2idx[tag]] = cnt 
    row += tmp_tag 
    new_df.loc[i] = row
    if i % 10 == 0: print("Status: ", i)
new_df
```
